```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from matplotlib import rc
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()

%matplotlib inline
```

```python
col = ['동','출생', '사망', '전입', '전출', '혼인', '이혼', '의원']
#col_1 = ['동','출생', '사망', '전입', '전출', '혼인', '이혼', '병원']
col_2 = ['동','출생', '사망', '전입', '전출', '혼인', '이혼', '치과의원']
s_kmean = add[col]
#s_kmean.sort_values(['동'], inplace=True)
```

```python
s_kmean
```

```python
mmdata = minmax.fit_transform(s_kmean.drop('동', axis=1))
mmdata = pd.DataFrame(mmdata, columns=s_kmean.drop('동',axis=1).columns)
kmeans = KMeans(n_clusters=3, random_state=1234).fit(mmdata)
s_kmean['cluster'] = kmeans.labels_
s_kmean
```

```python
add['cluster'] = s_kmean['cluster']
```

```python
import folium
m_6 = folium.Map(
    location = (sum(add['위도'])/len(add['위도']), sum(add['경도'])/len(add['경도']) ),
    tiles='cartodbdark_matter',
    zoom_start = 11)
for i in range(add.shape[0]):
    
    if add['cluster'][i] == 0:
        color = 'blue'
    elif add['cluster'][i] == 1:
        color = 'red'
    elif add['cluster'][i] == 2:
        color = 'green'
    else :
        pass
    c = add.iat[i,8],add.iat[i,9]
    folium.Circle(c, popup=add['cluster'][i], color = color).add_to(m_6)
```
